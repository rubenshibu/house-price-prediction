```python
def fill_na(x):
    return ((x['QS_ROOMS'] + x['QS_BATHROOM'] + x['QS_BEDROOM'])/3)
```

```python
df['QS_OVERALL'] = df.apply(lambda x:fill_na(x) if pd.isnull(x['QS_OVERALL'])else x['QS_OVERALL'],axis=1)
```
